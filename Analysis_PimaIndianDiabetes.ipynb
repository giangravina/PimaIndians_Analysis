8. Missing value is encoded as 0. How do you want to replace this with? Replace the 0's with the median with each variable

10. Draw histogram for predictor variables with vertical lines for 10 percentile, median and 90 percentile [Try subplots if you like]

```python
predictors = diabetes_df.drop('Outcome', axis=1)

fig, axs = plt.subplots(nrows=2, ncols=4, figsize=(15,8))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

for i, ax in enumerate(axs.flat):
    if i < predictors.shape[1]:
        col = predictors.columns[i]
        ax.hist(predictors[col], bins=10)
        ax.axvline(predictors[col].quantile(0.1), color='r', linestyle='dashed', linewidth=2)
        ax.axvline(predictors[col].median(), color='g', linestyle='dashed', linewidth=2)
        ax.axvline(predictors[col].quantile(0.9), color='r', linestyle='dashed', linewidth=2)
        ax.set_xlabel(col)
plt.show()
```

11. For each variable impute the value with the median if the value of a specific row lies beyond 10th and 90th percentile

```python
percentiles = diabetes_df.quantile([0.1, 0.9])

for col in diabetes_df.columns:
    if col != "Outcome":
        diabetes_df.loc[(diabetes_df[col] < percentiles.loc[0.1, col]) | (diabetes_df[col] > percentiles.loc[0.9, col]), col] = np.nan
        median = diabetes_df[col].median()
        diabetes_df[col].fillna(median, inplace=True)

print(diabetes_df.isna().sum())
```

16. Create a new variable segment1 with the following pseudo-code: if (Glucose <= 120 and Age <= 30) then segment= 1 else 0

```python
diabetes_df['segment1'] = 0

diabetes_df.loc[(diabetes_df['Glucose'] <= 120) & (diabetes_df['Age'] <= 30), 'segment1'] = 1
```

17. Draw a stack bar chart to show how the proportion of diabetic patients are different inside and outside the segment1. [Think: What does inside and outside mean]


```python
proportion_diabetic = diabetes_df.groupby('segment1')['Outcome'].mean()

plt.bar(proportion_diabetic.index, proportion_diabetic, label='Diabetic')
plt.bar(proportion_diabetic.index, 1-proportion_diabetic, bottom=proportion_diabetic, label='Non-Diabetic')
plt.xlabel('Segment1')
plt.ylabel('Proportion')
plt.legend()
plt.show()
```
